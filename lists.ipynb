## Delete Node in a Linked List

```
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def deleteNode(self, node):
        temp = node.next
        node.val = temp.val
        node.next = temp.next
        del temp
```

## Linked List Cycle

```
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def hasCycle(self, head: ListNode) -> bool:
        hare = head
        tortoise = head
        while hare and hare.next:
            hare = hare.next.next
            tortoise = tortoise.next
            if hare == tortoise:
                break
        else:
            return False
        return True
```

```
nums = [1, 2, 3, 4, 5, 6, 3]


from graphviz import Digraph
import pydotplus
from IPython.display import SVG


def plot_graph(ptr1, ptr2):
  dot = Digraph(graph_attr={'rankdir':'LR'}, 
                node_attr={'shape': 'point', 'width':'0.5'})
  for idx, val in enumerate(nums):
    node_name = 'index: ' + str(idx) + '\\nvalue: ' + str(val)
    color = 'black'
    if idx == ptr1:
      color = 'red'
    if idx == ptr2:
      color = 'green'
    if idx == ptr1 and idx == ptr2:
      color = 'yellow'
    dot.node(str(idx), label=node_name, group='main', color=color)
    dot.edge(str(idx), str(val))
    if idx < len(nums) -1:
      dot.edge(str(idx), str(idx+1), style='invis')

  graph = pydotplus.graph_from_dot_data(dot.source)
  display(SVG(graph.create_svg()))

plot_graph(-1, -1)
hare = nums[nums[0]]
tortoise = nums[0]
plot_graph(tortoise, hare)
while hare != tortoise:
  tortoise = nums[tortoise]
  hare = nums[nums[hare]]
  plot_graph(tortoise, hare)
```

## Intersection of Two Linked Lists

```
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> ListNode:
        ptrA, ptrB = headA, headB
        strikes = 0
        while strikes < 3:
            if ptrA == ptrB:
                break
            ptrA = ptrA.next
            ptrB = ptrB.next
            if not ptrA: 
                ptrA = headB
                strikes += 1
            if not ptrB: 
                ptrB = headA
                strikes += 1
        else:
            return None
        return ptrA
            
```

## Palindrome Linked List

```
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def isPalindrome(self, head: Optional[ListNode]) -> bool:
        def recursion(current_ptr):
            if not current_ptr: return head, True
            new_head, result = recursion(current_ptr.next)
            return new_head.next, result and (new_head.val == current_ptr.val)
        _, result = recursion(head)
        return result
```

## Reverse Linked List

```
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        ptrA = head
        ptrB = None
        
        while ptrA:
            ptrB, ptrA.next, ptrA = ptrA, ptrB, ptrA.next 

        return ptrB
```

```
class Node:
    def __init__(self, value, next=None):
        self.value = value
        self.next = next

head = Node(7)
head = Node(6, head)
head = Node(5, head)
head = Node(4, head)
head = Node(3, head)
head = Node(2, head)
head = Node(1, head)

def process(head):
  ptrA = head
  ptrB = None
  append_slide({"Old List": ptrA, "New List": ptrB})
  while ptrA:
    temp = ptrA
    ptrA = ptrA.next
    temp.next = ptrB
    ptrB = temp
    append_slide({"Old List": ptrA, "New List": ptrB})

process(head)
publish_slides()
```

## Odd Even Linked List

![](https://assets.leetcode.com/users/images/f84e9388-7ba0-43d3-8cbd-a2248cf1ca26_1611038094.2788172.png)

```
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def oddEvenList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head: return head;
        odd = head
        even = head.next
        evenHead = even
        while even and odd and even.next and odd.next:
            odd.next = even.next
            odd = odd.next
            even.next = odd.next
            even = even.next
        odd.next = evenHead
        return head
```

```
class Node:
    def __init__(self, value, next=None):
        self.value = value
        self.next = next

head = Node(7)
head = Node(6, head)
head = Node(5, head)
head = Node(4, head)
head = Node(3, head)
head = Node(2, head)
head = Node(1, head)
head = Node(0, head)


odd = head
even = head.next
evenHead = even
append_slide({"odd": odd, "oddHead": head, "evenHead": evenHead, "even": even})
while even and odd and even.next and odd.next:
  odd.next = even.next
  odd = odd.next
  even.next = odd.next
  even = even.next
  append_slide({"odd": odd, "oddHead": head, "evenHead": evenHead, "even": even})
odd.next = evenHead
append_slide({"odd": odd, "oddHead": head, "evenHead": evenHead, "even": even})

publish_slides()
```

## Remove Duplicates from Sorted List II

![](https://assets.leetcode.com/users/andriy111/image_1585794700.png)

```
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def deleteDuplicates(self, head: Optional[ListNode]) -> Optional[ListNode]:
        dummy = ListNode(val=-1, next=head)
        p, q = dummy, head
        while q:
            prev_val = q.val
            q = q.next
            if q and prev_val == q.val:
                continue
            if q != p.next.next:
                p.next = q
            else:
                p = p.next
        return dummy.next
```

```
class Node:
    def __init__(self, value, next=None):
        self.value = value
        self.next = next

head = Node(7)
head = Node(6, head)
head = Node(5, head)
head = Node(3, head)
head = Node(3, head)
head = Node(3, head)
head = Node(1, head)
head = Node(0, head)


dummy = Node(-1, head)
p, q = dummy, head
while q:
  prev_val = q.value
  q = q.next
  if q and prev_val == q.value:
    append_slide({"p": p, "q": q, "dummy": dummy})
    continue
  if q != p.next.next:
    p.next = q
  else:
    p = p.next
  append_slide({"p": p, "q": q, "dummy": dummy})

publish_slides()
```
